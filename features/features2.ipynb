```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import parselmouth
import soundfile as sf
import librosa
import os
import glob
import re
from feat_ext import *
from praat_class import Sinal
from tqdm import tqdm
```

```python
df = pd.read_csv("..\\dataset_key_v1.csv")
df
```

```python
cpps = []
slope = []
mediaF0 = []
desvioF0 = []
minimoF0 = []
maximoF0 = []
range_F0 =[]
mediaF1 = []
desvioF1 = []
minimoF1 = []
maximoF1 = []
range_F1 =[]
mediaF2 = []
desvioF2 = []
minimoF2 = []
maximoF2 = []
range_F2 =[]
mediaF3 = []
desvioF3 = []
minimoF3 = []
maximoF3 = []
range_F3 =[]
jitter = []
shimmer = []
hnr = []
gne = []

filename = df['caminho']
for i in tqdm(range(len(df))):
    a = Sinal(filename[i])
    a.carrega()
    a.pitch()
    a.jitter()
    a.shimmer()
    a.getFormants()
    a.getHNR()
    a.getGNE()
    a.getCPPS()
    a.getSlope()

    valor_med_f0 = a.meanF0
    valor_std_f0 = a.stdevF0
    valor_min_f0 = a.minF0
    valor_max_f0 = a.maxF0
    valor_ran_f0 = a.rangeF0
    valor_jitt = a.jitt
    valor_shimm = a.shimm
    valor_hnr = a.hnr
    valor_gne = a.gne
    valor_med_f1 = a.meanF1
    valor_std_f1 = a.stdevF1
    valor_min_f1 = a.minF1
    valor_max_f1 = a.maxF1
    valor_ran_f1 = a.rangeF1
    valor_med_f2 = a.meanF2
    valor_std_f2 = a.stdevF2
    valor_min_f2 = a.minF2
    valor_max_f2 = a.maxF2
    valor_ran_f2 = a.rangeF2
    valor_med_f3 = a.meanF3
    valor_std_f3 = a.stdevF3
    valor_min_f3 = a.minF3
    valor_max_f3 = a.maxF3
    valor_ran_f3 = a.rangeF3
    valor_cpps = a.cpps
    valor_slope = a.slope

    mediaF0.append(valor_med_f0)
    desvioF0.append(valor_std_f0)
    minimoF0.append(valor_min_f0)
    maximoF0.append(valor_max_f0)
    range_F0.append(valor_ran_f0)
    jitter.append(valor_jitt)
    shimmer.append(valor_shimm)
    hnr.append(valor_hnr)
    gne.append(valor_gne)
    mediaF1.append(valor_med_f1)
    desvioF1.append(valor_std_f1)
    minimoF1.append(valor_min_f1)
    maximoF1.append(valor_max_f1)
    range_F1.append(valor_ran_f1)
    mediaF2.append(valor_med_f2)
    desvioF2.append(valor_std_f2)
    minimoF2.append(valor_min_f2)
    maximoF2.append(valor_max_f2)
    range_F2.append(valor_ran_f2)
    mediaF3.append(valor_med_f3)
    desvioF3.append(valor_std_f3)
    minimoF3.append(valor_min_f3)
    maximoF3.append(valor_max_f3)
    range_F3.append(valor_ran_f3)
    cpps.append(valor_cpps)
    slope.append(valor_slope)
```

```python
df['mediaF0'] = mediaF0
df['desvioF0'] = desvioF0
df['minimoF0'] = minimoF0
df['maximoF0'] = maximoF0
df['rangeF0'] = range_F0
df['jitter'] = jitter
df['shimmer'] = shimmer
df['hnr'] = hnr
df['gne'] = gne
df['cpps'] = cpps
df['slope'] = slope
df['mediaF1'] = mediaF1
df['desvioF1'] = desvioF1
df['minimoF1'] = minimoF1
df['maximoF1'] = maximoF1
df['rangeF1'] = range_F1
df['mediaF2'] = mediaF2
df['desvioF2'] = desvioF2
df['minimoF2'] = minimoF2
df['maximoF2'] = maximoF2
df['rangeF2'] = range_F2
df['mediaF3'] = mediaF3
df['desvioF3'] = desvioF3
df['minimoF3'] = minimoF3
df['maximoF3'] = maximoF3
df['rangeF3'] = range_F3
df
```
