```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## carregando arquivos da classe PTL

```python
df_ptl = pd.read_pickle('/content/drive/MyDrive/proj_arq_2023/dataset_matrix_ptl_kay_v3.pkl')
df_ptl
```

### Dividindo PTL em treino, validação, e teste

- treino : 70%
- validação : 10%
- teste : 20%

```python
train_ptl, valid_ptl, test_ptl = \
              np.split(df_ptl.sample(frac=1, random_state=42),
                       [int(.7*len(df_ptl)), int(.8*len(df_ptl))])
```

```python
train_ptl
```

```python
valid_ptl
```

```python
test_ptl
```

```python
print("checando divisão PTL:")
print(r"PTL - Treino: {:.2f}%".format((len(train_ptl) / len(df_ptl))*100))
print(r"PTL - Validação: {:.2f}%".format((len(valid_ptl) / len(df_ptl))*100))
print(r"PTL - Teste: {:.2f}%".format((len(test_ptl) / len(df_ptl))*100))
```
