## changing data type: from float64 to float16

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
df_normal = pd.read_pickle("dataset_matrix_sdl_kay_v2.pkl")
df_normal
```

```python
type(df_normal)
```

```python
df_normal.dtypes
```

```python
df_normal.keys()
```

```python
colunas_float64_sdl = df_normal.select_dtypes(include=['float64']).columns
df_normal[colunas_float64_sdl] = df_normal[colunas_float64_sdl].astype('float16')
df_normal
```

```python
df_normal.dtypes
```

```python
colunas_int64_sdl = df_normal.select_dtypes(include=['int64']).columns
df_normal[colunas_int64_sdl] = df_normal[colunas_int64_sdl].astype('int16')
df_normal.dtypes
```

```python
df_normal.to_pickle("dataset_matrix_sdl_kay_v3.pkl")
df_normal.to_csv("dataset_matrix_sdl_kay_v3.csv")
```
