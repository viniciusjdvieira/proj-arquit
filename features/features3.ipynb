## obtendo dataframe para saudáveis

```python
from feat_ext import *

def get_todas_features(filename, data, order = 29):
    valor_medio_lpc, lpcs = get_lpc(data, order = 29)
    lpcs=np.array(lpcs)
    n_seg = np.shape(lpcs)[0]

    valor_medio_mfcc, mfccs = get_mfcc(data)
    mfccs = mfccs[:-2,:]
    valor_medio_d_mfcc, dmfccs = get_delta_mfcc(data)
    dmfccs = dmfccs[:-2,:]
    valor_medio_d_d_mfcc, ddmfccs = get_delta_delta_mfcc(data)
    ddmfccs = ddmfccs[:-2,:]

    mediaF0, desvioF0, minimoF0, maximoF0, range_F0 = get_matrix_f0(filename, n_frames = n_seg)
    matriz_f0 = np.array([mediaF0, desvioF0,minimoF0, maximoF0, range_F0]).T

    jitt = get_matrix_jitter(filename, n_frames = n_seg)
    matriz_jitter = np.array([jitt]).T

    shimm = get_matrix_shimmer(filename, n_frames = n_seg)
    matriz_shimmer = np.array([shimm]).T

    hnr = get_matrix_hnr(filename, n_frames = n_seg)
    matriz_hnr = np.array([hnr]).T

    gne = get_matrix_gne(filename, n_frames = n_seg)
    matriz_gne = np.array([gne]).T

    mediaF1, desvioF1, minimoF1, maximoF1, range_F1 = get_matrix_f1(filename, n_frames = n_seg)
    matriz_f1 = np.array([mediaF1, desvioF1,minimoF1, maximoF1, range_F1]).T

    mediaF2, desvioF2, minimoF2, maximoF2, range_F2 = get_matrix_f2(filename, n_frames = n_seg)
    matriz_f2 = np.array([mediaF2, desvioF2,minimoF2, maximoF2, range_F2]).T

    mediaF3, desvioF3, minimoF3, maximoF3, range_F3 = get_matrix_f3(filename, n_frames = n_seg)
    matriz_f3 = np.array([mediaF3, desvioF3,minimoF3, maximoF3, range_F3]).T

    slope = get_matrix_slope(filename, n_frames = n_seg)
    matriz_slope = np.array([slope]).T

    cpps = get_matrix_cpps(filename, n_frames = n_seg)
    matriz_cpps = np.array([cpps]).T

    matriz1 = np.concatenate([lpcs, mfccs, dmfccs, ddmfccs,
                           matriz_f0], axis=1)
    
    matriz2 = np.concatenate([matriz1, matriz_jitter, matriz_shimmer,
                              matriz_hnr, matriz_gne], axis=1)
    
    matriz3 = np.concatenate([matriz2, matriz_f1, matriz_f2,
                              matriz_f3, matriz_slope, matriz_cpps], axis=1)
    
    # lista de features:
    list_features = ['lpc1','lpc2','lpc3', 'lpc4','lpc5','lpc6','lpc7','lpc8',
                     'lpc9','lpc10','lpc11','lpc12','lpc13','lpc14','lpc15','lpc16',
                     'lpc17','lpc18','lpc19','lpc20','lpc21','lpc22','lpc23','lpc24',
                     'lpc25','lpc26','lpc27','lpc28','lpc29',
                     'mfcc1', 'mfcc2', 'mfcc3', 'mfcc4', 'mfcc5', 'mfcc6', 'mfcc7',
                     'mfcc8', 'mfcc9', 'mfcc10', 'mfcc11', 'mfcc12',
                     'dmfcc1', 'dmfcc2', 'dmfcc3', 'dmfcc4', 'dmfcc5', 'dmfcc6', 'dmfcc7',
                     'dmfcc8', 'dmfcc9', 'dmfcc10', 'dmfcc11', 'dmfcc12',
                     'ddmfcc1', 'ddmfcc2', 'ddmfcc3', 'ddmfcc4', 'ddmfcc5', 'ddmfcc6', 'ddmfcc7',
                     'ddmfcc8', 'ddmfcc9', 'ddmfcc10', 'ddmfcc11', 'ddmfcc12',
                     'mediaF0','desvioFO', 'minimoF0', 'maximoF0', 'rangeF0',
                     'jitter', 'shimmer', 'hnr', 'gne',
                     'mediaF1','desvioF1', 'minimoF1', 'maximoF1', 'rangeF1',
                     'mediaF2','desvioF2', 'minimoF2', 'maximoF2', 'rangeF2',
                     'mediaF3','desvioF3', 'minimoF3', 'maximoF3', 'rangeF3',
                     'slope', 'cpps']

    return matriz3, list_features

```

```python
from feat_ext import *

df_saudavel = pd.DataFrame()
filename = df_sdl['caminho']
paciente = df_sdl['arquivo_sem_wav']
classe = df_sdl['classe']
idx_na_classe = df_sdl['index_na_classe']
idx_class = df_sdl['index_classification']
for i in tqdm(range(len(df_sdl))):
    data, samplerate = sf.read(filename[i])
    valores, medidas = get_todas_features(filename[i], data)
    n_seg = np.shape(valores)[0]
    multi_name = [paciente[i]] * n_seg
    multi_class = [classe[i]] * n_seg
    multi_idxclass = [idx_na_classe[i]] * n_seg
    classification = [idx_class[i]] * n_seg
    
    if i == 0:
        df_saudavel['paciente'] = multi_name
        df_saudavel['classe'] = multi_class
        df_saudavel['idx_na_classe'] = multi_idxclass
        df_saudavel['classification'] = classification
        df_saudavel = pd.concat([df_saudavel, pd.DataFrame(valores, columns=medidas)], axis=1)

    else:
        df_normal = pd.DataFrame()
        df_normal['paciente'] = multi_name
        df_normal['classe'] = multi_class
        df_normal['idx_na_classe'] = multi_idxclass
        df_normal['classification'] = classification
        df_normal = pd.concat([df_normal, pd.DataFrame(valores, columns=medidas)], axis=1)

        df_saudavel = pd.concat([df_saudavel, df_normal], axis=0)

    
```
